```python
# IQR  : distance between the 3rd quartile and the 1st quartile
df["mpg"].quantile(0.75) - df["mpg"].quantile(0.25)
```

```python
# Upper Whisker: (Q3 + 1.5 * IQR) this point is the upper boundary before individual points are considered outliers.
#              22.8+7.375*1.5 =33.8625
# Lower Whisker: (Q1 – 1.5 * IQR)
df.boxplot(column="mpg",figsize=(6,6))

plt.text(x=0.75, y=22.800000, s="3rd Quartile")
plt.text(x=0.75, y=15.425000, s="1rd Quartile")
plt.text(x=1.1, y=18, s="IQR", rotation=90, size=20)
plt.text(x=1.2, y=15.425, s="<--------Box--------->", rotation=90, size=10)
plt.text(x=1.2, y=22.8, s="<-------------Whisker-------->", rotation=90, size=10)
plt.text(x=1.2, y=10.4, s="<--Whisker->", rotation=90, size=10)
plt.text(x=1.2, y=33.9, s="Outlier", rotation=90, size=10)

plt.text(x=0.8, y=19.200, s="Median")

plt.text(x=0.9, y=10.400000, s="Min")
plt.text(x=0.9, y=33.900000, s="Max")
plt.show()
```

```python
### normal distribution (Gaussian distribution)
# mean = median = mode    Skewness=0    Variance=std**2        
# z-score = (X-mean)/std
import scipy.stats  as stats
```

```python
# 1sd
prob_below_minus1 = stats.norm.cdf(x= -1, loc = 0, scale= 1)     
prob_over_1 = 1 - stats.norm.cdf(x= 1, loc = 0, scale= 1) 
between_prob = 1 - (prob_below_minus1 + prob_over_1)

print(prob_below_minus1, prob_over_1, between_prob)
```

```python
plt.figure(figsize=(5, 5))                                  
plt.fill_between(x=np.arange(-4,-1,0.01), 
                 y1= stats.norm.pdf(np.arange(-4,-1,0.01)) ,
                 facecolor='red',
                 alpha=0.35)

plt.fill_between(x=np.arange(1,4,0.01), 
                 y1= stats.norm.pdf(np.arange(1,4,0.01)) ,
                 facecolor='red',
                 alpha=0.35)

plt.fill_between(x=np.arange(-1,1,0.01), 
                 y1= stats.norm.pdf(np.arange(-1,1,0.01)) ,
                 facecolor='green',
                 alpha=0.35)

plt.text(x=-1.9, y=0.03, s= round(prob_below_minus1,3))
plt.text(x=-0.2, y=0.1, s= round(between_prob,3))
plt.text(x=1.3, y=0.03, s= round(prob_over_1,3))
plt.show()
```

```python
# 2sd
prob_below_minus2 = stats.norm.cdf(x= -2, loc = 0, scale= 1)     
prob_over_2 = 1 - stats.norm.cdf(x= 2, loc = 0, scale= 1) 
between_prob = 1 - (prob_below_minus2 + prob_over_2)

print(prob_below_minus2, prob_over_2, between_prob)
```

```python
plt.figure(figsize=(5, 5))                                  
plt.fill_between(x=np.arange(-4,-2,0.01), 
                 y1= stats.norm.pdf(np.arange(-4,-2,0.01)) ,
                 facecolor='red',
                 alpha=0.35)

plt.fill_between(x=np.arange(2,4,0.01), 
                 y1= stats.norm.pdf(np.arange(2,4,0.01)) ,
                 facecolor='red',
                 alpha=0.35)

plt.fill_between(x=np.arange(-2,2,0.01), 
                 y1= stats.norm.pdf(np.arange(-2,2,0.01)) ,
                 facecolor='green',
                 alpha=0.35)

plt.text(x=-2.8, y=0.03, s= round(prob_below_minus2,3))
plt.text(x=-0.6, y=0.1, s= round(between_prob,3))
plt.text(x=1.8, y=0.03, s= round(prob_over_2,3))
plt.show()
#Empirical Rule: About 68% of data falls within one standard deviation, 95% within two, and 99.7% within three.
```
