```python
# drop unused features
df = df.drop('Product' ,axis=1)
df
```

```python
# Convert Categorical Variable to Numeric
pd.get_dummies(df.Company, dtype=int)
```

```python
df = df.join(pd.get_dummies(df.Company, dtype=int))
df
```

```python
df = df.drop('Company', axis=1)
df
```

```python
df = df.join(pd.get_dummies(df.TypeName, dtype=int))
df = df.drop('TypeName',axis=1)
df
```

```python
df.ScreenResolution.str.split(' ').apply(lambda x: x[-1])
```

```python
df['ScreenResolution'] = df.ScreenResolution.str.split(' ').apply(lambda x: x[-1])
df
```

```python
df['Screen Width'] = df.ScreenResolution.str.split('x').apply(lambda x: x[0]).astype('int')
df['Screen Height'] = df.ScreenResolution.str.split('x').apply(lambda x: x[1]).astype('int')
df = df.drop('ScreenResolution' ,axis =1 )
df
```

```python
df.Cpu.value_counts()
```

```python
df['Cpu Brand'] = df.Cpu.str.split(' ').apply(lambda x: x[0])
df['Cpu Freq'] = df.Cpu.str.split(' ').apply(lambda x: x[-1])
df['Cpu Freq'] = df['Cpu Freq'].str[ :-3].astype('float')
df = df.drop('Cpu' ,axis = 1)
df
```

```python
df['Ram'] = df['Ram'].str[:-2].astype('int')
df
```

```python
df['Memory Amount'] = df.Memory.str.split(' ').apply(lambda x: x[0])
df['Memory Type'] = df.Memory.str.split(' ').apply(lambda x: x[1])

df
```

```python
df['Memory Amount'].value_counts()
```

```python
df = df.drop('Memory Type',axis=1)
df
```

```python
# func turn Memory Amount to int
def mem_to_MB(x) :
    if 'GB' in x:
        return float(x[ :x.find('GB')])*1000
    elif 'TB' in x :
        return float(x[ :x.find('TB')])*100000
```

```python
df['Memory Amount'].apply(mem_to_MB)
```

```python
df['Memory Amount'] = df['Memory Amount'].apply(mem_to_MB)
df = df.drop('Memory', axis=1)
df
```

```python
df.Weight.value_counts()
```

```python
df['Weight'] =df['Weight'].str[:-2].astype(float)
df
```

```python
df['Gpu Brand'] = df.Gpu.str.split(' ').apply(lambda x: x[0])
df = df.drop('Gpu',axis=1)
df
```

```python
df = df.join(pd.get_dummies(df.OpSys, dtype=int))
df = df.drop('OpSys',axis=1)
df
```

```python
cpu_categories = pd.get_dummies(df['Cpu Brand'] ,dtype=int)
cpu_categories.columns = [col+'_CPU' for col in cpu_categories.columns]
cpu_categories
```

```python
df = df.join(cpu_categories)
df = df.drop('Cpu Brand',axis = 1)
df
```

```python
gpu_categories = pd.get_dummies(df['Gpu Brand'] ,dtype=int)
gpu_categories.columns = [col+'_GPU' for col in gpu_categories.columns]
gpu_categories
```

```python
df = df.join(gpu_categories)
df = df.drop('Gpu Brand',axis = 1)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize = (18,16))
sns.heatmap(df.corr(), annot=True , cmap='YlGnBu')
```

```python
target_corr = df.corr()['Price_euros'].apply(abs).sort_values(ascending=False)
target_corr
```

```python
# Selected the top 20 features that have a high correlation with Price_euros
selected_features = list(target_corr[:21].index)
selected_features
```

```python
final_df = df[selected_features]
final_df
```

```python
plt.figure(figsize = (18,16))
sns.heatmap(final_df.corr(), annot=True , cmap='YlGnBu', linewidth=.5 )
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# ml part

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# 1. Split data
X,y = final_df.drop('Price_euros',axis = 1 ) , final_df['Price_euros']
X_train , X_test , y_train , y_test  =train_test_split(X,y , test_size = 0.2)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

## 2. Train
forest = RandomForestRegressor(n_estimators=100)
forest.fit(X_train_scaled , y_train)
```

```python
# 3. Score
forest.score(X_train_scaled , y_train)
```

```python
y_pred =  forest.predict(X_test_scaled)
```

```python
# 4. Evaluate
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(y_test, y_pred))
rmse
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_pred,y_test)
plt.xlabel('y_pred')
plt.ylabel('y_test (actual Price_euros)')
```
