```python
from sklearn.datasets import load_iris
data = load_iris()
data
```

```python
import numpy as np 
full_data = np.append(data.data , data.target.reshape(-1,1) , axis =1 )
full_data
```

```python
import pandas as pd
df = pd.DataFrame(full_data)
df
```

```python
df.columns = data.feature_names + ['Type']
df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

col = 'sepal length (cm)'
df[col].hist(bins = 15)
plt.suptitle(col)
plt.show()
```

```python
col = 'sepal width (cm)'
df[col].hist(bins = 15)
plt.suptitle(col)
plt.show()
```

```python
col = 'petal length (cm)'
df[col].hist(bins = 15)
plt.suptitle(col)
plt.show()
```

```python
col = 'petal width (cm)'
df[col].hist(bins = 15)
plt.suptitle(col)
plt.show()
```
