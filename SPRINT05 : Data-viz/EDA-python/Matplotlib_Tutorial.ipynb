```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# scatter plot 
X_data = np.random.random(size = 50)
y_data = np.random.random(size = 50)

plt.scatter(X_data, y_data, c='green', marker='*', s=100, alpha=0.4)
plt.show()
```

```python
# line plot
years = [2008 + x for x in range(16)]
weights = [81, 83, 85, 83, 80, 82, 84, 89, 79, 86, 81, 82, 87, 89, 82, 86]
print(years)
print(len(years))
print(len(weights))
```

```python
plt.plot(years, weights, c='b' ,lw=3 , linestyle='--')
plt.xlabel('years')
plt.ylabel('weights')
plt.title('years vs weights')
```

```python
# bar plot 
x= ['python', 'c++' , 'java']
y= [70, 30, 40]

plt.bar(x,y, color='r', align='edge', width=0.5 , edgecolor='blue' , lw=2 )
```

```python
# histogram
ages = np.random.normal(loc=20, scale=8.5, size=1000)

plt.hist(ages, bins=50);
```

```python
# pie chart
x = ['python', 'c++' , 'java', 'c#', 'Go']
votes = [50, 30, 40, 10, 20]
explodes = [0,0.2,0,0,0]

plt.pie(votes , labels=x, explode = explodes , autopct='%.2f%%', pctdistance = 0.8 , startangle=90);
```

```python
# boxplot
heights = np.random.normal(170,8,300)

plt.boxplot(heights);
```

```python
first = np.linspace(0,10,25)
second = np.linspace(10,100,25)
third = np.linspace(100,150,25)
fourth = np.linspace(150,200,25)

data = np.concatenate((first , second , third , fourth))
plt.boxplot(data);
```

```python
# customization
years = [2014 + x for x in range(8)]
income = [56 , 57, 59 ,63, 60 ,65 ,64 ,75]

income_ticks = list(range(50,81,2))
plt.plot(years , income)
plt.title('Income of sam by years ' , fontsize = 20 , fontname = 'monospace')
plt.xlabel('years' ,fontname = 'monospace')
plt.ylabel('Income' ,fontname = 'monospace')
plt.yticks(income_ticks, [f'{x}k USD' for x in income_ticks] , fontname = 'monospace')

plt.show()
```

```python
# multiple plot
x1 , y1 = np.random.random(100) , np.random.random(100)
x2 , y2 = np.arange(100) , np.arange(100) 

plt.figure(1)
plt.scatter(x1 , y1)

plt.figure(2)
plt.plot(x2 , y2)

plt.show()
```

```python
# subplot
style.use('classic')
x = np.arange(100)

fig ,axs = plt.subplots(2,2 ,figsize=(8, 6))

axs[0,0].plot(x , np.sin(x))

axs[0,1].plot(x , np.cos(x))

axs[1,0].plot(x , np.random.random(100))

axs[1,1].plot(x , np.log(x+1))
axs[1,1].set_title('log func')
axs[1,1].set_xlabel('xxxx')
axs[1,1].set_ylabel('yyyy')

fig.suptitle('4 plots')
plt.tight_layout()
#plt.savefig('plots.png' , dpi=300)
plt.show()
```

```python
# 3D plot
plt.style.use('default')
ax = plt.axes(projection='3d')

x= np.random.random(100)
y= np.random.random(100)
z= np.random.random(100)

ax.scatter(x,y,z)
ax.set_title('3D plot')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

```python
ax = plt.axes(projection ='3d')
 
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
c = x + y
ax.scatter(x, y, z, c = c)
```

```python
ax = plt.axes(projection ='3d')
x= np.arange(-5,5,0.1)
y= np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)
Z= np.sin(X)+np.cos(Y)
ax.plot_surface(X,Y,Z, cmap='Spectral')
plt.show()
```
