```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
tips
```

```python
sns.scatterplot(x= 'tip' , y= 'total_bill' ,data=tips , hue='day', size = 'size' , palette = 'viridis')
```

```python
sns.histplot(tips['tip'] , kde = True , bins = 25)
```

```python
sns.barplot(x = 'sex' , y = 'tip' , data =tips )
```

```python
sns.boxplot(x='day', y='total_bill' , data = tips , hue ='sex' , palette ='flare')
```

```python
sns.stripplot(x='day', y='tip', data=tips , hue='sex' , dodge =True)
```

```python
sns.jointplot(x='tip', y= 'total_bill' , data = tips , kind = 'reg' )
```

```python
sns.jointplot(x='tip', y= 'total_bill' , data = tips , kind = 'kde' ,shade =True )
```

```python
sns.jointplot(x='tip', y= 'total_bill' , data = tips , kind = 'hex' , cmap = 'rocket_r' )
```
