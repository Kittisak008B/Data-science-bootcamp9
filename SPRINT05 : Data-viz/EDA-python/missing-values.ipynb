```python
### forward fill , backward fill

import warnings
warnings.filterwarnings('ignore')
url ='https://pkgstore.datahub.io/core/oil-prices/brent-day_csv/data/f898de10e734eb3b019b17cd17e1841e/brent-day_csv.csv'
df_ukoil = pd.read_csv(url)
df_ukoil
```

```python
df_ukoil.info()
```

```python
df_oil=df_ukoil[(df_ukoil['Date'] >= '2015-01-05') & (df_ukoil['Date'] <= '2015-05-31')]
df_oil
```

```python
df_oil['Date'] = pd.to_datetime(df_oil['Date'])
df_oil
```

```python
df_oil.info()
```

```python
from_dt='20150101'
to_dt='20150531'
# to_dt=pd.Timestamp.now().strftime('%Y%m%d')
url=f'https://stooq.com/q/d/l/?s=^set&d1={from_dt}&d2={to_dt}&i=d'
df_set=pd.read_csv(url, parse_dates=['Date'], usecols=['Date', 'Close'])
df_set
```

```python
df_set.columns=['Date', 'SET_index']
df_set['Date'] = pd.to_datetime(df_set['Date'])
df_set
```

```python
df_set.info()
```

```python
# merge 2 df
df=pd.merge(df_oil, df_set, left_on='Date', right_on='Date', how='outer', indicator=True)
df
```

```python
df=df.sort_values('Date')
df
```

```python
df['_merge'].value_counts()
```

```python
df[df['_merge'] != 'both']
```

```python
df['Oil_f']=df['Brent Spot Price'].ffill() # forward fill
df['Oil_b']=df['Brent Spot Price'].bfill() # backward fill
df['SET_f']=df['SET_index'].ffill()
df['SET_b']=df['SET_index'].bfill()
df
```

```python
df[['Brent Spot Price', 'SET_index']].corr()
```

```python
df[['Oil_f', 'SET_f']].corr()
```

```python
df[['Oil_b', 'SET_b']].corr()
```

```python
sns.lineplot(data=df, x='Date', y='Oil_f', color='black')
ax2 = plt.twinx()
ax2.yaxis.label.set_color('gold')
ax2.tick_params(axis='y', colors='gold')
sns.lineplot(data=df, x='Date', y='SET_f', ax=ax2, color='gold')
```
