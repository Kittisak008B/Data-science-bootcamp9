```python
# perfomance comparison Iterrows vs Vectorization
import time

df = yf.download('^GSPC')
df
```

```python
start_time = time.time()

current_max = 0
for index,row in df.iterrows() :
    if current_max < row.High :
        current_max = row.High
        current_max_index = index

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
start_time = time.time()

df.nlargest(1,'High')

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# plot yearly returns of s&p500
df = yf.download('^GSPC' , start='2009-12-31')
```

```python
df['return'] = df.Close.pct_change() + 1
df.dropna(inplace=True)
df
```

```python
return_y = (df.resample('Y').apply('prod') -1 )['return']
return_y
```

```python
return_y.plot(kind='bar')
```

```python
df.resample('Y').last().Close.pct_change()
```
