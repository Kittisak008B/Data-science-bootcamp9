# The Star Wars API
## Get data from a public API and save it as a CSV file.

```python
from requests import get
from time import sleep
import pandas as pd

# People DataFrame 
url = "https://swapi.dev/api/people/"
a = []
for i in range(8) :
  index = i+1
  new_url = url + str(index)
  resp = get(new_url).json()
  print(resp['name'])
  a.append({
    "Name": resp["name"],
    "Gender": resp["gender"],
    "height" :resp["height"],
    "Birth_year": resp["birth_year"],
    "Homeworld": resp["homeworld"],
    "Films": resp["films"]
    })
  sleep(1)
df = pd.DataFrame(a)
df
```

```python
# Planets DataFrame 
b=[]
url2 = "https://swapi.dev/api/planets/"
for i in range(8) :
  index = i+1
  new_url = url2 + str(index)
  resp = get(new_url).json()
  print(resp['name'])
  b.append({
    "Homeworld": resp["name"],
    "terrain": resp["terrain"],
    "url": resp["url"]
    })
  sleep(1)
df2 = pd.DataFrame(b)
df2
```

```python
get("https://swapi.dev/api/films/1").json().keys()
```

```python
# Films DataFrame 
c=[]
url3 = "https://swapi.dev/api/films/"
for i in range(6) :
  index = i+1
  new_url = url3 + str(index)
  resp = get(new_url).json()
  print(resp['title'])
  c.append({
    "Film_title": resp["title"],
    "release_date": resp["release_date"],
    "director": resp["director"],
    "producer": resp["producer"],
    "url": resp["url"]
    })
  sleep(1)
df3 = pd.DataFrame(c)
df3
```

```python
result = pd.merge(df, df2 , how='left', left_on='Homeworld', right_on='url')
result
```

```python
#clean DataFrame
result.drop(['Homeworld_x', 'url','terrain'], axis=1, inplace=True)
result.rename(columns = {'Homeworld_y': 'Homeworld'}, inplace = True)

result
```
