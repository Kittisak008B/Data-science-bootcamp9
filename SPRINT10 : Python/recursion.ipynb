```python
# recursive functions can make the program take much longer to execute than using loops, but recursive functions are a wayofthinking.

#iterative
def factorial1(n) :
    if n==0 :
        return 1
    result = 1
    if n > 0 :
        for i in range(1,n+1) :
            result *= i
        return result
```

```python
for i in range(10) :
    print(factorial1(i))
```

```python
#recursive func

def factorial(n) :
    if n==0  :
        return 1
    if n >=1 :
        return n * factorial(n-1) 
```

```python
for i in range(10) :
    print(factorial(i))
```

```python
def sigma(n) :
    if n==0 :
        return 0
    else :
        return n + sigma(n-1)
```

```python
for i in range(10) :
    print(sigma(i))
```

```python
# count 0 in list
list_b = [1,0,2,3,0,0,4,0,8,0,10,0,1,0]
def func0(x) :
    count0 = 0
    for i in x :
        if i ==0 :
            count0 +=1
    return count0
```

```python
func0(list_b)
```

```python
# count 0   -recursion
def func0_recursion(x) :
    if len(x) == 0 :
        return 0
    if len(x) ==1 :
        if x[0] == 0 :
            return 1
        else :
            return 0
    if x[0]==0 :
        return 1 + func0_recursion(x[1:])
    else : 
        return 0 + func0_recursion(x[1:])
```

```python
func0_recursion(list_b)
```

```python
# find min in list 
list_c = [5,123,345,47,78,23,78,23,4]
def find_min(x) :
    minimum = x[0]
    for i in x :
        if i <= minimum :
            minimum = i
    return minimum
    
find_min(list_c)
```

```python
# find min in list   -recursion
def find_min_recursion(x) :
    if len(x) == 0 :
        return None
    if len(x) == 1 :
        return x[0]
    return min(x[0],find_min_recursion(x[1:]))
    
find_min_recursion(list_c)   
```

```python
# fibo 
def fibo(n) :
    a , b =0,1
    for i in range(0,n) :
        a,b = b,a+b
    return a
for i in range(10) :
    print(fibo(i))
```

```python
# fibo -recursion
# o(2^n)
import time
start_time = time.time()

def fibo_recursion(n) :
    if n <=1 :
        return n
    return fibo_recursion(n-1) + fibo_recursion(n-2)
    
print(fibo_recursion(40))

print("--- %s seconds ---" % (time.time() - start_time))
```
