```python
def getprices(ex, sym) :
    exchange = getattr(ccxt, ex)()
    df = pd.DataFrame(exchange.fetchOHLCV(sym ,limit=60))
    df.columns = ['Time','Open','High','Low','Close','Volume']
    df.set_index('Time' , inplace=True)
    df.index = pd.to_datetime(df.index , unit='ms',utc=True).map(lambda x:x.tz_convert('Asia/Bangkok'))
    df = df.astype(float)
    return df
```

```python
data = getprices('binance','BTC/USDT')
data
```
