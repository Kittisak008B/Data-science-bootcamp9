```python
from requests import get
from time import sleep
import pandas as pd
#Get a stand by an id
url = 'https://stand-by-me.herokuapp.com/api/v1/stands/'
a = []
for i in range(10) :
  index =  i+1
  new_url = url + str(index)
  resp = get(new_url).json()
  print(resp['name'])
  a.append({
    "Stand-Name": resp["name"],
    "japaneseName" :resp["japaneseName"],
    "standUser": resp["standUser"],
    "chapter": resp["chapter"],
    "abilities": resp["abilities"],
    "battlecry": resp["battlecry"]
    })
  sleep(1)
df = pd.DataFrame(a)
df["standUser"] = df["standUser"].astype(int)
df
```

```python
#Get a character by an id
url = 'https://stand-by-me.herokuapp.com/api/v1/characters/'
b = []
for i in range(26) :
  index =  i+1
  new_url = url + str(index)
  resp = get(new_url).json()
  print(resp['name'])
  b.append({
    "Character-Name": resp["name"],
    "nationality" :resp["nationality"],
    "catchphrase": resp["catchphrase"],
    "chapter": resp["chapter"],
    "isHuman": resp["isHuman"]
    })
  sleep(1)
df2 = pd.DataFrame(b)
index = pd.Index(range(1, 27, 1))
df2 = df2.set_index(index)
df2
```

```python
df3 = df2.drop(['nationality', 'catchphrase','chapter','isHuman'], axis=1)
df3
```

```python
result = pd.merge(df, df3 , how='left', left_on='standUser',right_index=True )
result.drop(['standUser'], axis=1,inplace = True)
result = result[['Stand-Name','japaneseName','Character-Name','abilities','battlecry','chapter']]
result.rename(columns = {'Character-Name': 'standUser'}, inplace = True)
result
```
